```python
import numpy as np
import pandas as pd
def fun(s):
    if s=='S':
        return 0
    elif s=='C':
        return 1
    else:
        return 2
def ch(s):
    if s=='male':
        return 0
    else:
        return 1
data=pd.read_csv("train_titanic.csv",sep=",")
del data["Name"]
del data["Ticket"]
del data["Cabin"]
data["embark"]=data.Embarked.apply(fun)
del data["Embarked"]
data["sx"]=data.Sex.apply(ch)
del data["Sex"]
data.Age.fillna(int(data.Age.mean()),inplace=True)
#data=np.loadtxt("train_titanic.csv",delimiter=",",skiprows=1)
y_train=data.iloc[:,5]
del data["Survived"]
data["c"]=1
x_train=data.values
x_train.shape
```

```python
def run(x,y):
    l=0.00001
    it=6000
    m=gd(x,y,l,it)
    print(m)
    return m
```

```python
def gd(x,y,l,it):
    m=np.zeros(x.shape[1])
    for i in range(it):
        m=step(x,y,l,m)
        print(i ,"cost :",cost(x,y,m))
    return m   
```

```python
import math
#b=math.exp()
def error(x,m):
    a=(x*m).sum()
    c=math.exp(-a)
    return (1/(1+c))
            
```

```python
def step(x,y,l,m):
    new=np.zeros(x.shape[1])
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            X=x[i]
            Y=y[i]
            h=error(X,m)
            new[j]+=(-1/x.shape[0])*(Y-h)*X[j]
            
        m=m-l*new
    return m   
            
```

```python
import math
#a=math.log()
def cost(x,y,m):
    total=0
    for i in range(x.shape[0]):
        Y=y[i]
        X=x[i]
        total+=(1/x.shape[0])*(-Y*(math.log(error(X,m)))-(1-Y)*(math.log(1-error(X,m))))
    return total   
    
    
```

```python
slope=run(x_train,y_train)
```
